```r
# install.packages('strucchange', repos = 'http://cran.us.r-project.org')
library('strucchange')
```

```r
oil.prices <- read.csv('./Data//oil_price.csv', colClasses = c('Date', 'character'))
```

```r
oil.prices <- oil.prices[oil.prices$VALUE != '.',]
oil.prices$VALUE <- as.numeric(oil.prices$VALUE)
str(oil.prices)
```

```r
plot(oil.prices, type = 'l')
```

```r
breaks.data <- breakpoints(oil.prices$VALUE ~ oil.prices$DATE)
```

```r
breaks.data$breakpoints
```

```r
pch.col <- sapply(1:400, function(x) {if (x %in% breaks.data$breakpoints) 'X' else '.'})
```

```r
plot(oil.prices, pch = pch.col)
```
